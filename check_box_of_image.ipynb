```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
from PIL import Image
from PIL import ImageDraw
import numpy as np
import os

def pil_draw_rect(img, c):
    draw = ImageDraw.Draw(img)
    draw.rectangle(((c[0], c[1]), (c[2], c[3])), outline=(100, 255, 255), width=2)
    return img


tg = input("type 0000_00 : ")
plt.figure(figsize=(1.5, 2))
texts = []

def bx_coord(line, height=480, width=640):
    # w, h = 640, 480
    n_s = line.split()
    n = [float(i) for i in n_s]
    box_f = (width*(n[1]-n[3]/2), height*(n[2]-n[4]/2), width*(n[1]+n[3]/2), height*(n[2]+n[4]/2))
    box = [int(i) for i in box_f]
    return box

for _ in range(1):
    plt.figure(figsize=(16, 18))
    #img = Image.open("/home/ai_competition6/2.testset/"+test_file_list[idx*2])
    img = Image.open("/home/ai_competition6/4.merge/"+tg[0:4] + "/" + tg+".jpg")
    
    w, h = img.size # width. height
    
    #f = open("/home/ai_competition6/2.testset/"+test_file_list[idx*2+1], "r")
    f = open("/home/ai_competition6/4.merge/"+tg[0:4] + "/" + tg+".txt", "r")
    text = f.readlines()
    texts.append(text)
    f.close()
    
    for l in text:
        img = pil_draw_rect(img, bx_coord(l, h, w))
        
    plt.imshow(img)
plt.show()

```
